# Data Cleaning and Analysis

```python
i = 0
for x in data['Mat']:
    if pd.isnull(x):
        inn = data.loc[i]['Inns']
        div = int(inn/2)
        data.at[i,'Mat'] = div
    i+=1
```

```python
# # remove the incorrect format
data['4s'] = data['4s'].astype(str).apply(lambda x: x.replace('+', '') if '+' in x else x)

# # Convert the column to numeric values
data['4s'] = pd.to_numeric(data['4s'], errors='coerce')
```

```python
data['BF'] = data['BF'].astype(str).str.replace('+','')
data['BF'] = pd.to_numeric(data['BF'], errors='coerce')
data['BF'] = data['BF'].fillna(method='ffill')
data['BF'] = data['BF'].astype('Int64')
```

```python
data['SR'] = data['SR'].replace('-','')
data['SR'] = pd.to_numeric(data['SR'], errors='coerce')

data['SR'] = data['SR'].fillna(method='ffill')
data['SR'] = data['SR'].astype(int)
data.loc[14]['SR']
```

```python
data.info()
```

```python
# now i will split a single colum into multiple
# data[['First Name', 'Last Name', 'Country']] = data['Player'].str.split(' ', 3, expand=True)
# Assuming 'data' is your DataFrame and 'Player' is the column to be split
data[['First Name', 'Last Name', 'Country']] = data['Player'].str.split(' ', n=2, expand=True)
```

```python
# remove bracket around country name
data.drop('Player', axis=1, inplace = True)
```

```python
data['Country'] = data['Country'].str.strip('12()')
```

```python
data['Country'] = data['Country'].str.replace('ICC/', '')
data
```

```python
index_columns = ['First Name', 'Last Name', 'Country']

# Set the specified columns as the index
data = data.set_index(index_columns)
```

```python
data = data.reset_index()
```

```python
# Distribution of players in term of scores, average and strike rate
fig = plt.figure(figsize=(50,50), dpi=100)
# sub_plot = fig.add_axes([0.1, 0.1, 0.2, 0.2])
fig,sub_plot=plt.subplots(nrows=1, ncols=3, figsize=(20,10), dpi=100)
sub_plot[0].hist(data['Runs'],  edgecolor='black', alpha=0.7)
sub_plot[1].hist(data['Ave'], edgecolor='black', alpha=0.7)
sub_plot[2].hist(data['SR'], edgecolor='black', alpha=0.7)
plt.tight_layout()
```

# I will find top 5 with the help of this graph

```python
# The below data is just for those who has best record than 95% of players
percentile_Runs = np.percentile(data['Runs'], 95)
percentile_Ave = np.percentile(data['Ave'], 95)
percentile_SR = np.percentile(data['SR'], 95)
percentile_SR
```

```python
record1 = data[(data['Runs'] >= percentile_Runs)]
record2 = data[(data['Ave'] >= percentile_Ave)]
record3 = data[(data['SR'] >= percentile_SR)]
record1
```

```python
# bar grapgh for comparison between top players
fig,subplot_comp = plt.subplots(nrows=1, ncols=3, figsize=(20,10), dpi=100)
subplot_comp[0].bar(record1['Last Name'], record1['Runs'], edgecolor='black', alpha=0.7, label='Top Runners')
subplot_comp[1].bar(record2['Last Name'], record2['Ave'], edgecolor='black', alpha=0.7, label='Top Average')
subplot_comp[2].bar(record3['Last Name'], record3['SR'], edgecolor='black', alpha=0.7, label='Top Strikers')

subplot_comp[0].set_title('Run Comparisom')
subplot_comp[1].set_title('Average Comparison')
subplot_comp[2].set_title('Strike Rate Comparison')
subplot_comp[0].set_xlabel('PLAYERS')
subplot_comp[1].set_xlabel('PLAYERS')
subplot_comp[2].set_xlabel('PLAYERS')
subplot_comp[0].set_ylabel('RUNS')
subplot_comp[1].set_ylabel('AVERAGE')
subplot_comp[2].set_ylabel('STRIKE RATE')

subplot_comp[0].legend()
subplot_comp[1].legend()
subplot_comp[2].legend()
plt.tight_layout()
```
