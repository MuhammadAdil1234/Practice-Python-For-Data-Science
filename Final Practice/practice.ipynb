```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arr = [1, 7, 33,37,31,39,40,41,42,45,30,36,150,850]
test = np.array(arr)
```

```python
# test.std()
# test.mean()
new1 = np.sort(test)
```

```python
np.median(test)
```

```python
plt.boxplot(test)
plt.show()
```

```python
# find IQR to identifying outlier
iqr3 = np.percentile(test, 75)
iqr1 = np.percentile(test, 25)
iqr = iqr3 - iqr1
iqr
```

```python
# for remove outlier
low_limit = iqr1 - 1.5*iqr
upper_limit = iqr3 + 1.5*iqr
for x in range(len(test)):
    if test[x] <= low_limit and test[x]>=upper_limit:
        new_arr = np.delete(test, x)
# another way
# # Create a boolean mask to filter out outliers
# mask = (test >= low_limit) & (test <= upper_limit)

# # Apply the mask to get a new array without outliers
# new_arr = test[mask]
```

```python
# new_arr.mean()
np.median(new_arr)
```

```python
plt.boxplot(new_arr)
plt.show()
```

```python
plt.hist(new_arr)
plt.show()
```

```python
import numpy as np

# Sample unsorted array
data = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5])


# Calculating the median (50th percentile) without sorting
median_value = np.percentile(data, 50)

print("Original unsorted array:", data)
print("Median without sorting:", median_value)
```

```python
import numpy as np

# Sample unsorted array
data = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5])

# Sorting the array
sorted_data = np.sort(data)

# Calculating the median (50th percentile) from the sorted array
median_value_sorted = np.percentile(sorted_data, 50)

print("Original unsorted array:", data)
print("Sorted array:", sorted_data)
print("Median from sorted array:", median_value_sorted)
```
