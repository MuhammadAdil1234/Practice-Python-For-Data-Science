```python
# some parameter for ploting important
fig = plt.figure(figsize=(3,2), dpi= 250) # figure size and its resolution
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y, label='X way')
axes.plot(y,x, label='Y way')
axes.legend(loc=0) # add legent the exact line in graph
# to save a plot in pc
fig.savefig('first.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# set the label axis and title
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('XY plot')
# also set range of plot on both x and y axis

axes.plot(x,y, label='Values', color='purple', linewidth = 3, linestyle='-', marker='o', markersize=10, markerfacecolor='yellow',
         markeredgewidth=2, markeredgecolor='blue')
axes.legend()
axes.set_xlim([0,5])
axes.set_ylim([10,15])
# add grid in the plot
axes.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```
